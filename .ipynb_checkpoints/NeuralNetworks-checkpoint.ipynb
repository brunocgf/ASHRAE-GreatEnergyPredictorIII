```python
import torch
import torch.nn as nn
import torchvision
from torch.autograd import Variable
from torch.utils.data import DataLoader
from torchvision.transforms import ToTensor
from ASHRAE_NN import NN_1
from ASHRAE_NN import NN_2
import matplotlib.pyplot as plt
import ASHRAE_data_prep
import numpy as np
```

```python
X_train, y_train = ASHRAE_data_prep.train_boost3()
```

```python
input_size = 7
num_classes = 1
num_epochs = 10
bs = 100
lr = 0.001
```

```python
X_train.shape
```

```python
X_train = X_train.to_numpy()
```

```python
y_train = y_train.to_numpy()
```

```python
train_data = []

for i in range(20216100):
   train_data.append([X_train[i], y_train[i]])

trainloader = DataLoader(train_data, shuffle=True, batch_size=bs)
```

```python
model = NN_1(input_size, num_classes)
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(),lr=lr)
```

```python
for epoch in range(num_epochs):

  for i, data in enumerate(trainloader,0):
    
    inputs, labels = data

    # Propagacion para adelante
    output = model(inputs.float())
    loss = loss_function(output, labels.float())
    # Propagcion para atras y paso de optimizacion

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (i+1) % 100 == 0:
      print ('Epoca: {}/{}, Paso: {}/{}, Perdida: {:.5f}'.format(epoch+1,num_epochs, i+1, len(trainloader), loss.item()))
```

```python
with torch.no_grad():
    y_train_pred = model(Variable(torch.from_numpy(X_train)).float()).data.numpy()
```
